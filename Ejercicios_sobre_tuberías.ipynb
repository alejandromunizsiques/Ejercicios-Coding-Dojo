# Librerias

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.neighbors import KNeighborsClassifier
```

# Montaje de dataset

```python
df = pd.read_csv('/content/drive/MyDrive/Data Science Coding Dojo/IntroduccioÃÅn al Machine Learning/abalone.data')
df
```

# Selecion de features y target

```python
X = df.drop(["M"],axis=1)
y = df["M"]
```

# Division de entrenamiento

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state= 1)
```

# Creacion de Pipeline

```python
pipe = make_pipeline(StandardScaler(), PCA(n_components = 8), KNeighborsClassifier(n_neighbors=15))
pipe.fit(X_train, y_train)
```

# Evaluacion de modelo

```python
print('Training accuracy:', pipe.score(X_train, y_train))
print('Testing accuracy:', pipe.score(X_test, y_test))
```
